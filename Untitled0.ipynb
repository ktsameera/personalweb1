```python
from sklearn.linear_model import LinearRegression
import numpy as nm
import matplotlib.pyplot as plt
x=nm.array([[2000],[2500],[3000],[3500],[4000]])
y=nm.array([15,20,25,30,35])
model=LinearRegression()
model.fit(x,y)
pr=model.predict([[3200]])
print("price prediction:",pr)
plt.scatter(x, y, color='blue')  # actual data
plt.plot(x, model.predict(x))    # regression line
plt.scatter(3200, pr, color='red')  # prediction point

plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Price Prediction")
plt.show()
```
